```python
# Derece, dakika, saniye formatındaki koordinatları decimal formata dönüştür
def dms_to_decimal(coord_str):
    degrees, minutes, seconds = map(float, coord_str.split(':'))
    decimal_coord = degrees + minutes/60 + seconds/3600
    return decimal_coord
```

```python
# Enlem ve boylam sütunlarını güncelle, 
df['KUZEY'] = df['KUZEY'].apply(dms_to_decimal) #enlem
df['DOĞU'] = df['DOĞU'].apply(dms_to_decimal) #boylam
df.head()
```

```python
# Mevsimi belirleyen bir fonksiyon tanımlama
def determine_season(date):
    if date.month in [12, 1, 2]:
        return "Kış" 
    elif date.month in [3, 4, 5]:
        return "İlkbahar"  
    elif date.month in [6, 7, 8]:
        return "Yaz"  
    else:
        return "Sonbahar"  

# Mevsim sütununu oluşturma
df['MEVSİM'] = df['BAŞLAMA TARİHİ VE SAATİ'].apply(determine_season)

# Sonuçları kontrol etmek için
print(df[['BAŞLAMA TARİHİ VE SAATİ', 'MEVSİM']].head())
```

```python
import pandas as pd
import re

# Gerekli sözlüklerin tanımlanması
tree_codes = {
    "Çz": "Kızılçam", "Çk": "Karaçam", "Çs": "Sarıçam", "G": "Göknar", "L": "Ladin", "S": "Sedir", "Ar": "Ardıç", "Çf": "Fıstık Çamı",
    "Sr": "Servi", "P": "Porsuk", "Çh": "Halep çamı", "Çm": "Sahil çamı", "Çr": "Radiata çamı", "D": "Duglaz", "An": "Andız",
    "Çt": "Taeda çamı", "Çe": "Elderica çamı", "Mlz": "Melez(Larix)", "Mzı": "Mazı", "Di": "Diğer ibreli",
    "Kn": "Kayın", "M": "Meşe", "Gn": "Gürgen", "Kz": "Kızılağaç", "Kv": "Kavak", "Ks": "Kestane", "Dş": "Dişbudak", "Ih": "Ihlamur",
    "Ak": "Akağaç", "Ka": "Karaağaç", "Ky": "Kayacık", "Çn": "Cınar", "Ok": "Okaliptus", "Sğ": "Sığla", "Fn": "Fındık", "Sö": "Söğüt",
    "H": "Huş", "Df": "Defne", "Ş": "Şimşir", "O": "Orman gülü", "Cv": "Ceviz", "Zy": "Yabani zeytin", "Mp": "Palamut meşesi",
    "Ms": "Saplı meşe", "Mz": "Sapsız meşe", "Mcr": "Macar meşesi", "Mt": "Tüylü meşe", "Mm": "Mazı meşesi", "Ml": "Saçlı meşe",
    "Mr": "Pırnal meşesi", "Bm": "Badem", "Ko": "Koca yemiş", "Mak": "Maki", "IL": "Ilgın", "Kb": "Kıbrıs akasyası", "Ya": "Yalancı akasya",
    "Ah": "Ahlat", "Yki": "Yabani kiraz(mahlep)", "Üv": "Üvez", "Dy": "Diğer yapraklı", "Mk": "Kasnak meşesi", "Mkr": "Kermes meşesi",
    "Hr": "Harnup", "Çi": "Çitlembik", "Mg": "Menengiç", "Sz": "Sakız", "Sl": "Sandal", "As": "Akçakesme", "Hm": "Hurma", "Ylk": "Yalankoz",
    "Ayl": "Aylantus", "İğ": "İğde", "Af": "Antep fıstığı", "Al": "Alıç", "Kzl": "Kızılcık", "Mı": "Istranca meşesi", "Mn": "Makedonya meşesi",
    "Mln": "Lübnan meşesi"
}
non_tree_codes = {
    "Ag0": "Ağaçlandırmaya Hazırlanmış Açık Alan", "OT": "Ağaçsız Orman Toprağı", "Yy": "Yayla Alanları", "OT-T": "Taşlı Ağaçsız Orman Toprağı",
    "OT-E": "Erozyonlu Ağaçsız Orman Toprağı", "F": "Orman Fidanlığı", "T": "Kayalık, Taşlık", "Ku": "Kum", "Bt": "Bataklık, Sazlık",
    "Dp": "Orman Deposu", "YDZ": "Yangın Durdurma Zonu", "Oc": "Ocak", "Ts": "İzin Verilmiş Alanlar", "OY": "Otoyol ve Bölünmüş Yol Şevleri ile Birlikte",
    "EH": "Elektrik,Doğalgaz vb. Hatları", "ZA": "Ziyaret Alanları"
}
non_forest_codes = {
    "KDa": "Kadastro Dışı Ağaçlık Alan", "Su": "Göl,Bent,Baraj,Nehir", "Me": "Mera", "2B": "2B ile Orman Dışı Bırakılmış",
    "İs": "İskan Alanı", "Mzl": "Devlet Orman Alanı Dışındaki Mezarlıklar", "Z": "Tarım Arazisi", "Z-İs": "Ziraat ve İskan Alanları",
    "Maki": "Maki", "DA": "Devlet Ormanı Dışındaki Diğer Alanlar"
}
canopy_cover = {
    "B": "Boşluklu kapalı", "1": "Az Kapalı", "2": "Orta Kapalı", "3": "Tam Kapalı"
}
age_stages = {
    "a": "Gençlik ve sıklık (çap 7.9 cm’ye kadar)", "b": "Sırıklık ve direklik (çap 8-19.9 cm)",
    "c": "İnce ağaçlık (çap 20-35.9 cm)", "d": "Orta ağaçlık (çap 36-51.9 cm)",
    "e": "Kalın ağaçlık (çap 52 cm ve üzeri)", "0": "Boş veya boşaltılmış orman alanı"
}

# Sütundaki birden çok bileşeni ayrıştırma
def parse_multiple_components(value):
    components = re.findall(r'(\d+)-\(([^-]+)-([^)]+)?\)', str(value))
    if components:
        parsed = [{'Bölme No': comp[0], 'Meşcere Tipi': comp[1]} for comp in components]
        return parsed
    return None

# Meşcere detaylarını işleme
def assign_meşcere_details(row, tree_codes, non_tree_codes, non_forest_codes, canopy_cover, age_stages):
    meşcere_tipi_sembolü = row['MEŞCERE TİPİ SEMBOLÜ']
    if pd.isna(meşcere_tipi_sembolü):
        return pd.Series(["Bilinmiyor"] * 4, index=['MEŞCERE KARŞILIĞI', 'ORMAN ALAN BİLGİSİ', 'TEPE KAPALILIĞI', 'GELİŞİM ÇAĞI'])
    meşcere_tipleri = meşcere_tipi_sembolü.split(", ")
    karşılıklar, alanlar, kapalılıklar, yaşlar = [], [], [], []
    for sembol in meşcere_tipleri:
        karşılık = alan = kapalılık = yaş = "Bilinmiyor"
        for code, desc in tree_codes.items():
            if code in sembol:
                karşılık, alan = desc, "Ağaçlı Ormanlık Alan"
                break
        for code, desc in non_tree_codes.items():
            if code in sembol:
                karşılık, alan = desc, "Ağaçsız Ormanlık Alan"
                break
        for code, desc in non_forest_codes.items():
            if code in sembol:
                karşılık, alan = desc, "Orman Dışı Alan"
                break
        for code, desc in canopy_cover.items():
            if code in sembol:
                kapalılık = desc
                break
        for code, desc in age_stages.items():
            if code in sembol:
                yaş = desc
                break
        karşılıklar.append(karşılık)
        alanlar.append(alan)
        kapalılıklar.append(kapalılık)
        yaşlar.append(yaş)
    return pd.Series([", ".join(karşılıklar), ", ".join(alanlar), ", ".join(kapalılıklar), ", ".join(yaşlar)],
                     index=['MEŞCERE KARŞILIĞI', 'ORMAN ALAN BİLGİSİ', 'TEPE KAPALILIĞI', 'GELİŞİM ÇAĞI'])

# Meşcere detaylarını işleme
df['PARSED_COMPONENTS'] = df['BÖLME NO - MEŞCERE - AĞAÇ TÜRÜ\n'].dropna().apply(parse_multiple_components)
parsed_columns = df['PARSED_COMPONENTS'].apply(lambda row: {
    'BÖLME NO': ', '.join([comp['Bölme No'] for comp in row]) if row else None,
    'MEŞCERE TİPİ SEMBOLÜ': ', '.join([comp['Meşcere Tipi'] for comp in row]) if row else None
}).apply(pd.Series)
df = pd.concat([df, parsed_columns], axis=1)
df.drop(columns=['PARSED_COMPONENTS'], inplace=True)
df[['MEŞCERE KARŞILIĞI', 'ORMAN ALAN BİLGİSİ', 'TEPE KAPALILIĞI', 'GELİŞİM ÇAĞI']] = df.apply(
    assign_meşcere_details, axis=1,
    args=(tree_codes, non_tree_codes, non_forest_codes, canopy_cover, age_stages)
)
df.head()
```


##### Boşluklu Kapalı: Yangın Riski: Düşük -> Tepe kapalılığı az olan ormanlarda, yangın genellikle yüzey yangını olarak kalır ve bu tür yangınlar kontrol altına alınması daha kolaydır. Çünkü alevler genellikle yer seviyesindeki bitki örtüsünde kalır ve ağaçların tepe kısmına sıçrama olasılığı düşüktür.
##### Az Kapalı: Yangın Riski: Orta 
##### Orta Kapalı: Yangın Riski: Orta- Yüksek 
##### Tam Kapalı: Yangın Riski: En Yüksek -> Tepe kapalılığının yüksek olduğu ormanlarda ise yangın riski artar. Ağaçların tepe kısmındaki dalların ve yaprakların birbirine yakın olması, yangının ağaçların tepelerine sıçrayarak hızla yayılmasına neden olabilir. 

```python
# Tepe Kapalılığı için sıralama  (tepe kapalılığı ne kadar azsa, yangın riskinin daha az)
tepe_kapalılığı_mapping = {
    "Boşluklu kapalı": 1, # Yangın riski:Düşük
    "Az Kapalı": 2, # Yangın Riski: Orta
    "Orta Kapalı": 3, # Yangın Riski: Orta-Yüksek
    "Tam Kapalı": 4,  #Yangın Riski: En yüksek
    "Bilinmiyor": 0
}

# Kodlama fonksiyonu
def ordinal_encode_tepe_kapalılığı_max(value):
    if pd.isna(value):  # Eksik değer kontrolü
        return 0  # Eksik veriyi "Bilinmiyor" olarak kabul et
    values_list = value.split(", ")  # Virgül ile ayrılmış değerleri ayır
    encoded_values = [tepe_kapalılığı_mapping.get(v.strip(), 0) for v in values_list]
    valid_values = [val for val in encoded_values if val != 0]  # "Bilinmiyor" (0) değerlerini filtrele
    if valid_values:
        return max(valid_values)  # En yüksek değeri döndür
    else:
        return 0  # Tüm değerler "Bilinmiyor" ise veya geçersizse 0 döndür

# Tepe Kapalılığı sütununu encode et
df['TEPE KAPALILIĞI_ENCODED'] = df['TEPE KAPALILIĞI'].apply(ordinal_encode_tepe_kapalılığı_max)

# Sonuçları incele
print(df[['TEPE KAPALILIĞI', 'TEPE KAPALILIĞI_ENCODED']].head())
```

```python
# Gelişim Çağı için sıralama
age_stages_mapping = {
    "Boş veya boşaltılmış orman alanı": 6, #Yangın Riski: Çok Yüksek
    "Gençlik ve sıklık (çap 7.9 cm’ye kadar)": 5, #Yangın Riski: Yüksek
    "Sırıklık ve direklik (çap 8-19.9 cm)": 4,
    "İnce ağaçlık (çap 20-35.9 cm)": 3,
    "Orta ağaçlık (çap 36-51.9 cm)": 2,
    "Kalın ağaçlık (çap 52 cm ve üzeri)": 1, #Yangın Riski: Düşük
    "Bilinmiyor": 0
}

# Kodlama fonksiyonu
def ordinal_encode_gelisim_cagi(value):
    if pd.isna(value):  # Eksik değer kontrolü
        return 0  # Bilinmeyen durumu
    values_list = value.split(", ")  # Virgül ile ayrılmış değerleri ayır
    encoded_values = [age_stages_mapping.get(v.strip(), -1) for v in values_list]
    valid_values = [val for val in encoded_values if val != 0]  # Bilinmeyenleri filtrele
    if valid_values:
        return max(valid_values)  # En yüksek değeri döndür
    else:
        return 0  # Tüm değerler "Bilinmiyor" ise veya geçersizse

# Gelişim Çağı sütununu encode et
df['GELİŞİM_ÇAĞI_ENCODED'] = df['GELİŞİM ÇAĞI'].apply(ordinal_encode_gelisim_cagi)

# Sonuçları incele
print(df[['GELİŞİM ÇAĞI', 'GELİŞİM_ÇAĞI_ENCODED']].head())
```
